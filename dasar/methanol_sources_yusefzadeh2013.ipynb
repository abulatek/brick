### Get 36 GHz methanol maser catalog from Yusef-Zadeh et al. 2013

```python
from astroquery.vizier import Vizier
import astropy.units as u
import astropy.coordinates as coord

Vizier.ROW_LIMIT = -1

# Get the Yusef-Zadeh 2013 catalog
masers = Vizier.get_catalogs("J/ApJ/764/L19")[0]
masers.write('methanol_masers_yusefzadeh2013.csv', overwrite = True)
masers.write('methanol_masers_yusefzadeh2013.fits', overwrite = True)
masers
```

### Get coordinates of those masers

```python
masers_glon = masers['GLON']
masers_glat = masers['GLAT']
```

### Get just one methanol line and plot catalog on top

```python
fn_95 = "/orange/adamginsburg/brick_alma_linesurvey/2019.1.00092.S/imaging_results/source_ab_95_spw25_clean_2sigma_n50000_masked_3sigma_pbmask0p18.image"
methanol_95 = methanol_lines[methanol_lines['freq_spw'] == '95_spw25']
methanol_95
```

```python
cube_95 = SpectralCube.read(fn_95, use_dask=True).with_spectral_unit(u.GHz).to(u.K)

slc_95 = cube_95[:1].subcube_slices_from_mask(cube_95[:1].mask)

cube_95 = cube_95[:,slc_95[1],slc_95[2]]

cube_95.beam_threshold = 10000

restf_95 = u.Quantity(methanol_95[0]['Frequency (GHz)'], u.GHz)
if (cube_95.spectral_axis.min() < restf_95) and (cube_95.spectral_axis.max() > restf_95):
    scube_95 = cube_95.with_spectral_unit(u.km/u.s, rest_value=restf_95, velocity_convention='radio').spectral_slab(vmin, vmax)
    m0_95 = scube_95.moment0(axis=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np

SM_SIZE = 10
MD_SIZE = 12
LG_SIZE = 14

plt.rc('font', size = MD_SIZE)          # controls default text sizes
plt.rc('axes', titlesize = LG_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize = MD_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize = MD_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize = MD_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize = SM_SIZE)    # legend fontsize
plt.rc('figure', titlesize = LG_SIZE)   # fontsize of the figure title

ax1 = plt.subplot(projection = m0_95.wcs)

im1 = ax1.imshow(m0_95.value, origin = 'lower', cmap = 'Greys_r')
ax1.autoscale(False)
ax1.scatter(masers_glon, masers_glat, transform = ax1.get_transform('galactic'), 
            marker = 'x', s=50, edgecolor='c', facecolor='c')

cbar1 = plt.colorbar(mappable = im1, ax = ax1, label = r"Integrated intensity [K km s$^{-1}$]", fraction=0.046, pad=0.04)

for ax in [ax1]:
    ax.tick_params(direction = 'in')
    ax.xaxis.set_tick_params(labelbottom = False)
    ax.yaxis.set_tick_params(labelleft = False)
    ax.set_ylabel('Declination')
    ax.set_xlabel('Right ascension')
    
ax1.set_xlim()
    
ax1.set_title("Known 36 GHz methanol masers on 95 GHz emission")
    
plt.rcParams.update({
    "figure.facecolor":  'w',
    "axes.facecolor":    'w',
    "savefig.facecolor": 'w'
})
# print(f'{basedir}/methanol_comparison_{round(vmax.value)}kms.pdf')
# plt.savefig(f'{basedir}/methanol_comparison_{round(vmax.value)}kms.pdf')
# plt.savefig(f'{basedir}/methanol_comparison_{round(vmax.value)}kms.png', dpi = 250, bbox_inches = 'tight')
# Also save to paper directory
# plt.savefig(f'/blue/adamginsburg/abulatek/brick/dasar/632e0742965a6e88270d73d7/figures/methanol_comparison_{round(vmax.value)}kms.pdf', bbox_inches = 'tight')
# plt.savefig(f'/blue/adamginsburg/abulatek/brick/dasar/632e0742965a6e88270d73d7/figures/methanol_comparison_{round(vmax.value)}kms.png', dpi = 250, bbox_inches = 'tight')
```

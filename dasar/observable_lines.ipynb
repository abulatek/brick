```python
import numpy as np
import matplotlib.pyplot as plt

restfs = [107.013831, 156.602395, 205.791270] # in GHz
```

```python
!which latex
```

```python
redshifts = np.array([0, 0.5, 1, 2, 3, 4, 5])

obsfs = []
for restf in restfs:
    for zz in redshifts:
        obsf = restf/(1+zz)
        obsfs.append(obsf)
obsfs = np.array(obsfs)
```

```python
obsfs_arr = np.reshape(obsfs, (len(restfs), redshifts.shape[0]))
obsfs_arr
```

```python
SM_SIZE = 12
MD_SIZE = 14
LG_SIZE = 18

plt.rc('font', size = MD_SIZE)          # controls default text sizes
plt.rc('axes', titlesize = LG_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize = MD_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize = MD_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize = MD_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize = SM_SIZE)    # legend fontsize
plt.rc('figure', titlesize = LG_SIZE)   # fontsize of the figure title

plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
})
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 5))

linestyles = ['-','--',':']

# Show dasar lines
for nn, obsf in enumerate(obsfs_arr):
    ax.plot(redshifts, obsf, label = f"CH$_3$OH {obsf[0]:.3f} GHz", linestyle = linestyles[nn])

# Make red section to show unobservable region
ax.axhspan(50.5, 70., alpha=0.1, color='red', label = 'Atmospheric lines (not observable)') # water bands
    
# Limit to observable bandwidth
plt.ylim(1.2, 116) # in GHz

plt.yticks([1.2, 20, 40, 60, 80, 100, 116])

plt.xlabel("Redshift, $z$")
plt.ylabel("Observable frequency [GHz]")
    
plt.grid(alpha = 0.5)
    
plt.legend()

# Make sure plot backgrounds are right for poster
ax.set(facecolor = "white")
fig.patch.set_alpha(0.0)
    
# plt.savefig('632e0742965a6e88270d73d7/figures/observable_lines.pdf', bbox_inches = 'tight')
# plt.savefig('632e0742965a6e88270d73d7/figures/observable_lines.png', dpi = 200, bbox_inches = 'tight')

plt.show()
```

```python
# Make a basic continuum image: take the median along the spectral axis somewhere without a whole lot of emission
from spectral_cube import SpectralCube

results = '/blue/adamginsburg/abulatek/brick/symlinks/imaging_results/smoothed_cubes_K/'
ind = '93_spw27'
fn = 'source_ab_'+ind+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.image.commonbeam.K.fits'
cube = SpectralCube.read(results+fn, use_dask = True)
print(cube)

simple_continuum = cube.median(axis=0)
```

```python
simple_continuum.quicklook()
```

```python
import matplotlib.pyplot as plt

ax1 = plt.subplot(111, projection = simple_continuum.wcs)
im1 = ax1.imshow(simple_continuum.value, origin = 'lower', cmap = 'Greys_r')
cbar1 = plt.colorbar(im1, ax=ax1)
cbar1.set_label(f"Continuum intensity [{simple_continuum.unit.to_string(format = 'latex_inline')}]")
ax1.set_ylabel('Declination')
ax1.set_xlabel('Right ascension')
```

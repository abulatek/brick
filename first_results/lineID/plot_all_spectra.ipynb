### Plot all unlabeled spectra

```python
# Create list of filenames and plot spectra
i = 0
for freq_spw in freq_spws:
    # Get max spectrum
    max_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.max.fits'
    max_sp = OneDSpectrum.from_hdu(fits.open(max_fn)).to(u.K)
    max_sp_py = pyspeckit.Spectrum.from_hdu(max_sp.hdu)
#     sp.baseline()
    max_sp_py.data -= np.median(max_sp_py.data)
    max_sp_py.xarr.convert_to_unit('GHz')

    # Get mean spectrum
    mean_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.mean.fits'
    mean_sp = OneDSpectrum.from_hdu(fits.open(mean_fn)).to(u.K)
    mean_sp_py = pyspeckit.Spectrum.from_hdu(mean_sp.hdu)
    mean_sp_py.data -= np.median(mean_sp_py.data)
    mean_sp_py.xarr.convert_to_unit('GHz')
    
    print(i,': ',freq_spw)
    
    # Initalize figure
    fig = plt.figure(1, figsize = (20, 4), tight_layout = True)
    
    # Plot max spectrum
    plt.subplot(121)
    max_sp_py.plotter(figure = fig)
    plt.title(f'{freq_spw} max')
    
    # Plot mean spectrum
    plt.subplot(122)
    mean_sp_py.plotter(figure = fig)
    plt.title(f'{freq_spw} mean')
    
    plt.savefig(f'/blue/adamginsburg/abulatek/brick/first_results/lineID/plots/unlabeled/{freq_spw}.pdf')
    plt.show()
    
    i += 1
```

### Plot individual spectra as we identify lines in them

```python
freq_spws_ids = [
'87_spw25',
'102_spw106'
]
```

```python
from astropy.io import ascii

every_line = ascii.read('ACES_lines_brick_updated_20210512.csv', format='csv')
every_line_by_spectrum = every_line.group_by(['freq_spw','Spectrum type'])

max_ids_87_spw25_L = every_line_by_spectrum.groups[2] # Need to find a better way
mean_ids_87_spw25_L = every_line_by_spectrum.groups[3] # to automate this part
max_ids_102_spw106_L = every_line_by_spectrum.groups[0]
mean_ids_102_spw106_L = every_line_by_spectrum.groups[1]

curated_max_ids = [max_ids_87_spw25_L,max_ids_102_spw106_L]
curated_mean_ids = [mean_ids_87_spw25_L,mean_ids_102_spw106_L]

# Create list of filenames and plot spectra
i = 0
for freq_spw in freq_spws_ids[0:2]:
    # Pick out ids
    max_ids = curated_max_ids[i]
    mean_ids = curated_mean_ids[i]
    
    # Get max spectrum
    max_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.max.fits'
    max_sp = OneDSpectrum.from_hdu(fits.open(max_fn)).to(u.K)
    max_sp_py = pyspeckit.Spectrum.from_hdu(max_sp.hdu)
#     sp.baseline()
    max_sp_py.data -= np.median(max_sp_py.data)
    max_sp_py.xarr.convert_to_unit('GHz')

    # Get mean spectrum
    mean_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.mean.fits'
    mean_sp = OneDSpectrum.from_hdu(fits.open(mean_fn)).to(u.K)
    mean_sp_py = pyspeckit.Spectrum.from_hdu(mean_sp.hdu)
    mean_sp_py.data -= np.median(mean_sp_py.data)
    mean_sp_py.xarr.convert_to_unit('GHz')
    
    print(i,': ',freq_spw)
    
    # Initalize figure
    fig = plt.figure(1, figsize = (20, 4), tight_layout = True)
    
    # Plot max spectrum
    plt.subplot(121)
    max_sp_py.plotter(figure = fig)
    max_sp_py.plotter.line_ids(line_names=max_ids['Species'], # ChemicalName 
                                line_xvals=max_ids['Frequency (GHz)'] - (max_ids['Frequency (GHz)']*(max_ids['Velocity (km/s)']/299792.)), 
                                xval_units='GHz', 
                                plot_kwargs={'color':'silver'})
    plt.title(f'{freq_spw} max')
    
    # Plot mean spectrum
    plt.subplot(122)
    mean_sp_py.plotter(figure = fig)
    mean_sp_py.plotter.line_ids(line_names=mean_ids['Species'], # ChemicalName 
                                line_xvals=mean_ids['Frequency (GHz)'] - (mean_ids['Frequency (GHz)']*(mean_ids['Velocity (km/s)']/299792.)), 
                                xval_units='GHz', 
                                plot_kwargs={'color':'silver'})
    plt.title(f'{freq_spw} mean')
    
    plt.savefig(f'/blue/adamginsburg/abulatek/brick/first_results/lineID/plots/labeled/{freq_spw}_labels.pdf')
    plt.show()
    
    i += 1
```

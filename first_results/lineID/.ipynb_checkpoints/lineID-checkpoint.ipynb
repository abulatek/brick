```python
freq_spw = '87_spw25'
max_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.max.fits'
max_sp = OneDSpectrum.from_hdu(fits.open(max_fn)).to(u.K)
max_sp_py = pyspeckit.Spectrum.from_hdu(max_sp.hdu)
max_sp_py.data
plt.figure(figsize = (10, 5))
plt.plot(max_sp_py.xarr, max_sp_py.data, linestyle = '-', linewidth = 1, drawstyle = 'steps-mid')
plt.xlim(np.min(max_sp_py.xarr.value), np.max(max_sp_py.xarr.value))
```

```python
# Create list of filenames and plot spectra
i = 0
for freq_spw in freq_spws[0:1]:
    # Get max spectrum
    max_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.max.fits'
    max_sp = OneDSpectrum.from_hdu(fits.open(max_fn)).to(u.K)
    max_sp_py = pyspeckit.Spectrum.from_hdu(max_sp.hdu)
#     sp.baseline()
    max_sp_py.data -= np.median(max_sp_py.data)

    # Get mean spectrum
    mean_fn = results+'source_ab_'+freq_spw+'_clean_2sigma_n50000_masked_3sigma_pbmask0p18.mean.fits'
    mean_sp = OneDSpectrum.from_hdu(fits.open(mean_fn)).to(u.K)
    mean_sp_py = pyspeckit.Spectrum.from_hdu(mean_sp.hdu)
    mean_sp_py.data -= np.median(mean_sp_py.data)
    
    print(i,': ',freq_spw)
    
    # Iniitalize figure
    plt.figure(1, figsize = (20, 4), tight_layout = True)
    
    # Plot max spectrum
    plt.subplot(121)
    plt.plot(max_sp_py.xarr/(10**9), max_sp_py.data, linestyle = '-', linewidth = 0.5, color = 'k', 
             drawstyle = 'steps-mid')
    plt.xlabel('Center frequency (GHz)')
    plt.ylabel('Brightness Temperature (K)')
    plt.xlim(np.min(max_sp_py.xarr.value/(10**9)), np.max(max_sp_py.xarr.value/(10**9)))
    plt.title(f'{freq_spw} max')
    
    # Plot mean spectrum
    plt.subplot(122)
    plt.plot(mean_sp_py.xarr/(10**9), mean_sp_py.data, linestyle = '-', linewidth = 0.5, color = 'k', 
             drawstyle = 'steps-mid')
    plt.xlabel('Center frequency (GHz)')
    plt.ylabel('Brightness Temperature (K)')
    plt.xlim(np.min(mean_sp_py.xarr.value/(10**9)), np.max(mean_sp_py.xarr.value/(10**9)))
    plt.title(f'{freq_spw} mean')
    
    plt.savefig(f'/blue/adamginsburg/abulatek/brick/first_results/lineID/plots/{freq_spw}.pdf')
    plt.show()
    
    i += 1
```
